```
import mxnet as mx
from bert_embedding import BertEmbedding
ctx = mx.gpu(0)
bert = BertEmbedding(ctx=ctx,model='bert_24_1024_16',dataset_name='book_corpus_wiki_en_cased') #Intializing BERT model and dataset corpus


inp=[]
with open("dataset.txt","r") as f:
  for sen in f:
    inp.append(sen.strip()) # creating a list of all the sentences in dataset.txt
  f.close()

f = open('bert_300d.txt', 'w')
dit={}
a=0
b=1000
c=1
print(len(inp))
while(b<=len(inp)):
  result = bert(inp[a:b]) # passing 1000 sentences at a time to the model to get the bert embeddings
  for i in range(len(result)):
    for j in range(len(result[i][0])):
      if result[i][0][j] not in dit:
        dit[result[i][0][j]]=1
        f.write(str(result[i][0][j])+" ")
        for y in result[i][1][j][:300]: # Consider the embedding size before writing into the embeddings file 
          f.write(str(y)+" ")
        f.write("\n")
  if b>=len(inp):
    break
  print(a,b)
  a+=1000
  if b+1000<=len(inp):
    b+=1000
  else:
    b=len(inp)
  print(c)
  c+=1
f.close()
```
